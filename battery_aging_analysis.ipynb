```python
import pandas as pd

# Load the metadata CSV file
df_metadata = pd.read_csv('/content/nasa_battery_data/cleaned_dataset/metadata.csv')

# Filter only the 'impedance' tests where Re and Rct are available
impedance_data = df_metadata[df_metadata['type'] == 'impedance']

# Check the first few rows to confirm
print(impedance_data.head())
```

```python
# Drop rows where Re or Rct are NaN (optional)
impedance_data_clean = impedance_data.dropna(subset=['Re', 'Rct'])

# Optionally, sort by test_id or start_time to show aging progression
impedance_data_clean = impedance_data_clean.sort_values(by='test_id')

# Check the cleaned data
print(impedance_data_clean.head())
```

```python
import plotly.express as px

# Plot Electrolyte Resistance (Re) over test_id (aging process)
fig_re = px.line(impedance_data_clean, x='test_id', y='Re', title='Electrolyte Resistance (Re) Over Aging (Test IDs)')
fig_re.show()

# Plot Charge Transfer Resistance (Rct) over test_id (aging process)
fig_rct = px.line(impedance_data_clean, x='test_id', y='Rct', title='Charge Transfer Resistance (Rct) Over Aging (Test IDs)')
fig_rct.show()
```

```python
# Plot both Re and Rct over test_id on the same graph
fig_all = px.line(impedance_data_clean, x='test_id', y=['Re', 'Rct'],
                  title='Electrolyte and Charge Transfer Resistance (Re & Rct) Over Aging')
fig_all.show()

```

```python
import numpy as np

# Function to extract the magnitude of a complex number (if applicable)
def extract_magnitude(value):
    try:
        # If the value is a string representation of a complex number, convert it
        if isinstance(value, str):
            value = value.strip("()")
            real, imag = value.split('-') if '-' in value else value.split('+')
            real = float(real)
            imag = float(imag.replace('j', ''))
            return np.sqrt(real**2 + imag**2)
        return float(value)  # If already a float, return as is
    except Exception as e:
        return np.nan  # If any error occurs, return NaN

# Apply the function to convert complex numbers in the columns
impedance_data_clean['Re'] = impedance_data_clean['Re'].apply(extract_magnitude)
impedance_data_clean['Rct'] = impedance_data_clean['Rct'].apply(extract_magnitude)

# Now, drop rows where Re or Rct are NaN (optional)
impedance_data_clean = impedance_data_clean.dropna(subset=['Re', 'Rct'])

```

```python
# Apply rolling mean to smooth the data (5-point window)
impedance_data_clean['Re_smooth'] = impedance_data_clean['Re'].rolling(window=5).mean()
impedance_data_clean['Rct_smooth'] = impedance_data_clean['Rct'].rolling(window=5).mean()

# Plot the smoothed data
import plotly.express as px

# Plot Smoothed Electrolyte Resistance (Re)
fig_re_smooth = px.line(impedance_data_clean, x='test_id', y='Re_smooth', title='Smoothed Electrolyte Resistance (Re) Over Aging')
fig_re_smooth.show()

# Plot Smoothed Charge Transfer Resistance (Rct)
fig_rct_smooth = px.line(impedance_data_clean, x='test_id', y='Rct_smooth', title='Smoothed Charge Transfer Resistance (Rct) Over Aging')
fig_rct_smooth.show()
```
